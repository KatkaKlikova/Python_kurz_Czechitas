```python
import pandas as pd
import matplotlib.pyplot as plt
```

Stáhni si soubor platy_2021_02.csv s informacemi o platech v softwarové firmě (stejný soubor, jako v předchozím úkolu).

```python
platy = pd.read_csv("platy_2021_02.csv")
platy.head()
```

Načti si tato data do tabulky a vytvoř histogram. Nastav vhodně hranice skupin histogramu (parametr bins), aby byl graf přehledný a snadno interpretovatelný.

```python
platy.hist(column="plat",
           bins=7,
           range=(26000, 65000),
            rwidth=0.75,
            facecolor="g"
           )

plt.title("Platy zaměstnanců 02/2021\n", fontweight="bold")
plt.xlabel("Plat v Kč")
plt.xticks(range(25000, 70000, 10000), [25000, "málo", "ujde to", "dost dobrý!", 65000])

```

Vrať se k práci se souborem temperature.csv, který obsahuje informace o průměrné teplotě v různých městech v listopadu 2017.

```python
teploty = pd.read_csv("temperature.csv")
teploty.head()
```

Vytvoř tabulku, která bude obsahovat údaje o teplotě za města Helsinki, Miami Beach a Tokyo.

```python
tri_mesta = teploty[(teploty["City"]=="Helsinki") | (teploty["City"]=="Miami Beach")|(teploty["City"]=="Tokyo")]
tri_mesta_orez = tri_mesta[["City", "AvgTemperature"]]
tri_mesta_orez.head()
```

Vytvoř krabicový graf a porovnej rozsah teplot v těchto městech.

```python
tri_mesta[["City", "AvgTemperature"]].groupby("City").plot(kind="box", )
```

```python
tri_mesta[["City", "AvgTemperature"]].groupby("City").boxplot()
```
